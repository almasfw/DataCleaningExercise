```python
import pandas as pd
import numpy as np
import re
```

```python
nominees = pd.read_csv("oscars.csv", encoding='latin1')
```

```python
birthPlaces = list(nominees["birthplace"])
for i in range(len(birthPlaces)):
    birthPlaces[i] = re.findall(r'\s?([\w\s]+)$', birthPlaces[i])
print(birthPlaces)
```

```python
for i in range(len(birthPlaces)):
    nominees.loc[[i], "birthplace"] = birthPlaces[i]
    #nominees = nominees.replace(nominees["birthplace"][i], birthPlaces[i][0])
```

```python
nonUS = nominees["birthplace"]
nonUS = list(dict.fromkeys(nonUS)) #remove duplication
for i in range(len(nonUS)):
    if (len(nonUS[i]) <= 2) | (nonUS[i] == "New York City"):
        nonUS.pop(i)
nonUS = list(nonUS)
nonUS
```

```python
baris_sebelum = nominees.shape[0]
print("Jumlah nominee sebelumnya adalah {}.".format(baris_sebelum))

nominees = nominees.drop(nominees["birthplace"].isin(nonUS))

baris_sesudah = nominees.shape[0]
print("Jumlah nominee setelah seleksi {}.".format(baris_sesudah))

print("{} nominee yang dihapus!".format(baris_sebelum-baris_sesudah))
```
